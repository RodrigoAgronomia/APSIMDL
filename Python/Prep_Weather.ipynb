```python
import pyreadr
import numpy as np
import pandas as pd
```

```python
def df2array(df):
    m,n = len(df.columns.levels[0]), len(df.columns.levels[1])
    arr = df.values.reshape(-1,m,n)
    return(arr)
```

```python
wdf = pyreadr.read_r('/home/germanm2/vr_value/Data/met_files/weather_historic_dt.rds')[None]
wdf
```

```python
wdf['year_day'] = 1e3*wdf.year + wdf.day
```

```python
cls = ["radn", "maxt", "mint", "rain", "dayl"]
wdfw = wdf.pivot(index='id_10', columns= 'year_day', values = cls)
wdfw
```

```python
# Dim1: Locations, Dim2: Variables, Dim3: Days

wdfwnp = df2array(wdfw).astype('float32')
```

```python
np.save('/home/rodrigo7/Documents/APSIMDL/weather.npy', wdfwnp)
```
